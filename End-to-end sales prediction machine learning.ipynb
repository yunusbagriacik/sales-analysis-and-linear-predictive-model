```python
sns.barplot(data=grouped_ceyrek_2017, x='Çeyrek', y='Satuş Tutarı')

# Annotate each bar with its value
for index, row in grouped_ceyrek_2017.iterrows():
    plt.text(row.name, row['Satuş Tutarı'], str(row['Satuş Tutarı']), ha='center')

plt.show()
```

```python
# Calculate the percentage increase compared to the previous Çeyrek
grouped_ceyrek_2017['Percent Increase'] = grouped_ceyrek_2017['Satuş Tutarı'].pct_change() * 100
grouped_ceyrek_2017['Percent Increase']
```

```python
# Plot the percentage increase
sns.barplot(data=grouped_ceyrek_2017, x='Çeyrek', y='Percent Increase')
plt.ylabel('Percentage Increase')
plt.title('Percentage Increase by Previous Year')
plt.show()
```

```python
# Check for missing values
missing_cols = X.columns[X.isnull().any()]

# Drop columns with missing values
X = X.drop(missing_cols, axis=1)

# Calculate variance and select non-zero variance columns
variance = X.var(axis=0, numeric_only=True, ddof=0)
variance_df = pd.DataFrame(variance, index=X.columns, columns=['Variance'])
non_zero_variance_cols = variance_df.index[variance_df['Variance'] != 0]

# Print information for debugging
print("Missing Columns:")
print(missing_cols)
print("Variance:")
print(variance_df)
print("Non-Zero Variance Columns:")
print(non_zero_variance_cols)

# Update X with non-zero variance columns
X = X[non_zero_variance_cols]

# Print the modified DataFrame
print("Modified X:")
print(X)
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.preprocessing import StandardScaler
import numpy as np
from sklearn.model_selection import train_test_split


SEED = 42

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = SEED)

# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_scaled = scaler.fit_transform(X)

# Apply the same scaling to the test set
X_test_scaled = scaler.transform(X_test)
```

 ### Linear Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error



linear_model = LinearRegression()
linear_model.fit(X_train_scaled, y_train)
y_pred_li = linear_model.predict(X_test_scaled)

print(mean_squared_error(y_test, y_pred_li))
```

### Lasso Regression



```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error
import numpy as np

alphas = np.arange(0, 1, 0.01)

best_alpha = None
lowest_mse = float('inf')

for alpha in alphas:
    lasso_model = Lasso(alpha=alpha)
    lasso_model.fit(X_train, y_train)
    y_pred_la = lasso_model.predict(X_test)

    mse = mean_squared_error(y_test, y_pred_la)
    if mse < lowest_mse:
        lowest_mse = mse
        best_alpha = alpha

print("Best Alpha:", best_alpha)
print("Lowest Mean Squared Error:", lowest_mse)
```

### Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, mean_absolute_error
from sklearn.model_selection import cross_val_score

ridge_model = Ridge(alpha=0.1)

# Fit the Ridge model on the scaled training data
ridge_model.fit(X_train_scaled, y_train)

# Predict on the scaled test data
y_pred_r = ridge_model.predict(X_test_scaled)

# Calculate mean squared error
mse = mean_squared_error(y_pred_r, y_test)
print('MSE:', mse)

rmse = np.sqrt(mse)
print('RMSE:', rmse)

# Perform k-fold cross-validation
k = 5  # number of folds
cv_scores = cross_val_score(ridge_model, X_scaled, y, cv=k, scoring='neg_mean_squared_error')

# Convert negative MSE scores to positive
cv_scores = -cv_scores

# Calculate the mean and standard deviation of cross-validation scores
mean_cv_score = cv_scores.mean()
std_cv_score = cv_scores.std()

print("Cross-Validation Mean Squared Error:", mean_cv_score)
print("Cross-Validation Standard Deviation:", std_cv_score)

# Calculate MAE
mae = mean_absolute_error(y_test, y_pred_r)
print("MAE:", mae)
```

```python
outlier_threshold = 2000
plt.scatter(range(len(y_test)), y_test, color='red', alpha = 0.5, label='Real Values')
plt.scatter(range(len(y_test)), y_pred_r, color='blue', alpha = 0.25, label='Predicted Values')
plt.axhline(y=outlier_threshold, color='black', linestyle='--', label='Outlier Threshold')
plt.axhline(y=-outlier_threshold, color='black', linestyle='--')
plt.xlabel('Data Point')
plt.ylabel('Value')
plt.title('Real and Predicted Values with Outliers')
plt.legend()
plt.show()
```
